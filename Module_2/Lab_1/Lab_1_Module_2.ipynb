#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

x_col = "Steps"
y_columns = ["Minutes Lightly Active", "Minutes Fairly Active", "Minutes Very Active"]


for y_col in y_columns:

    ax = plt.gca() # gca = current axes or chart
    ax.scatter(fitbit[x_col], fitbit[y_col])
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.set_title("{} vs {}".format(x_col, y_col))
    
    plt.show()
```

```python
# all elements in one chart

Lightly = fitbit['Minutes Lightly Active']
Fairly = fitbit['Minutes Fairly Active']
Very = fitbit['Minutes Very Active']
Steps = fitbit['Steps']

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# define the scatter plots
ax.scatter(Steps, Lightly, label = 'Lightly', color = 'r')
ax.scatter(Steps, Fairly, label = 'Fairly', color = 'g')
ax.scatter(Steps, Very, label = 'Very', color = 'blue')
ax.set_title('Scatter Plot')
ax.set_xlabel('Steps')
ax.set_ylabel('Minutes')

# add legends
ax.legend()

plt.show()
```
