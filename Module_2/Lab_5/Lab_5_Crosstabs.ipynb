### 2. Use crosstabs to show the number of moovies per country

```python
movies.columns
```

```python
Country = pd.crosstab(movies.country, movies.movie_title, margins = True, margins_name='Count').reset_index()
Country[['country', 'Count']].sort_values('Count')
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
dir_gen = pd.crosstab([movies.director_name, movies.genres], movies.country)
dir_gen
```

### 4. Average duration of filmes grouped by color

```python
duration = pd.crosstab(movies.duration, movies.color)
movies.groupby('color')['duration'].mean()
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
movies.groupby('director_name')['director_facebook_likes'].mean().sort_values(ascending=False)
```
