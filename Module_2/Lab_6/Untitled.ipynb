```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
x
```

```python

plt.plot(x, y, label ='linear')
plt.plot(x, z, label ='square')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(10, 3))

# create chart 1
ax1.plot(x,y, color = 'r', ls = '--')

# create chart 2
ax2.plot(x,z, color = 'g')

plt.show()
```
