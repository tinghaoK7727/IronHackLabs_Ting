Challenge 4: Given an array of distinct integers, return all the possible permutations. You can return the answer in any order. The lenght of array and the integers are given by user and may vary

```python
import numpy as np

N = int(input('Enter the total number of integers in your list:'))
M = int(input('Enter the maximum integer you wish to have in the list:'))
randomlist = np.random.randint(0,M+1,N)
p = list(itertools.permutations(randomlist))
print(randomlist)
print(p)
```

Challenge 5: Use intertools library to solve the following problem: Six books are lying on a table in front of you. How many ways can you arrange the books, considering both the left-to-right order of the books and whether theyre set with the front cover facing up or down?

```python
Books = [1, 2, 3, 4, 5, 6]
Cover = [1, 0]

B_permutations = itertools.permutations(Books)
C_permutations = itertools.product(Cover, repeat = len(Books))

Final = itertools.product(B_permutations, C_permutations)
# need to add 'list' to operate on the itertools result
print(len(list(Final)))
```
