We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

# sub 1 - aggregate amount spent to each customer ID
Cus_Spent = df.groupby('CustomerID', as_index = False).agg({'amount_spent':'sum'})
Cus_Spent.columns = ['CustomerID','Cus_Spent']
# merge two tables
df_merged = pd.merge(df, Cus_Spent, on = 'CustomerID')
df_merged.head()
```

```python
# sub 2 - finding the percentiles of "VIP" and "Preferred"
# DataFrame.quantile(q=0.5, axis=0, numeric_only=True, interpolation='linear')

# VIP: amount spent above 5840.182
Cus_Spent.quantile(0.95)

# Preferred: amount spent between 1661.64 and 5840.182
Cus_Spent.quantile([0.75, 0.95])
```

```python
# sub 3 - labeling the customers as "VIP" and "Preferred"

df_merged['Cus_Status'] = np.where(df_merged['Cus_Spent'] > 5840.182, 'VIP', 
                                   (np.where(df_merged['Cus_Spent'] > 1661.64, 'Preferred Customer', '')))
df_merged.head()
```

```python
df_final = df_merged[['CustomerID', 'Country', 'Cus_Status']].drop_duplicates()
df_final.head()
```

```python
# VIP or Preferred

VIP_Pre = df_final[(df_final['Cus_Status']=='VIP') | (df_final['Cus_Status'] =='Preferred Customer')]
VIP_Pre
```
