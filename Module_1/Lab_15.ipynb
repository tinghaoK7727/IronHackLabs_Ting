Question 1: Find the questions and answers of last month.

```python
from datetime import date
from datetime import datetime
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=datetime(2022,2,1), todate=datetime(2022,2,28))
answers = SITE.fetch('answers', fromdate=datetime(2022,2,1), todate=datetime(2022,2,28))
questions
answers
```

Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'.

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=datetime(2022,3,2), min=5, tagged='python', sort='votes')
questions
```

Question 3: Find the answers with id 6784 and 6473.

```python
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', ids=[6784, 6473])
answers
```
