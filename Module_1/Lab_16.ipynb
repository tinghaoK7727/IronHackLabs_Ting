#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
table = soup.find_all('table',{'class':'wikitable sortable'})[0]
rows = table.find_all('tr')
rows = [row.text.strip().split('\n') for row in rows]
colnames = rows[0]
data = rows[1:]
df = pd.DataFrame(data, columns = colnames)
df.head(10)
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
movies = soup.find_all('td', {'class':'titleColumn'})
rows = [row.text.strip().split('\n') for row in movies]
df = pd.DataFrame(rows, columns = ['Number', 'Movie', 'Year'])
df
````

#### Find the book name, price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html, 'lxml')
data_1 = soup.find_all('h3')
list1 = [i.text.strip().split('\n') for i in data_1] # a list of lists
print(list1)
list_1 = [''.join(i) for i in list1] # a list of strings
print(list_1)
data_2 = soup.find_all('div', {'class':'product_price'})
list_2 = [j.text.strip().split('\n')[0] for j in data_2] # a list of strings
print(list_2)
list_3 = [k.text.strip().split('\n')[4] for k in data_2] # a list of strings
print(list_3)
zip_list = zip(list_1, list_2, list_3)
df = pd.DataFrame(zip_list, columns = ['Name', 'Price', 'Availability'])
df
````
