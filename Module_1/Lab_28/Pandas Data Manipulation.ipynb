```python
import pandas as pd
import numpy as np
```

```python
# upload the file by pd.read_csv

file = pd.read_csv(r'/Users/tinghaokao/Desktop/vehicles/vehicles.csv')
file
```

```python
file.columns
```

```python
# reorder the columns

column_order = ['Year','Make','Model','Vehicle Class',
                'Transmission','Drivetrain','Fuel Type',
                'Cylinders','Engine Displacement','Fuel Barrels/Year',
                'City MPG','Highway MPG','Combined MPG',
                'CO2 Emission Grams/Mile','Fuel Cost/Year']


file = file[column_order]
file.head()
```

```python
# to filter the data by certain criteria

filtered = file[(file['Fuel Type'] == 'Regular') & (file['Cylinders']>=6)]
filtered.head()
```

```python
# group the date 

avg_mpg = file.groupby('Make', as_index=False).agg({'Combined MPG':'mean'})
avg_mpg.columns = ['Make', 'Avg_MPG']
```

```python
# to merge the data

file = pd.merge(file, avg_mpg, on='Make')
file.head(10)
```
